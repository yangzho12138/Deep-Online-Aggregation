```python
from tpch_polars import run_thread
import numpy as np
import pandas as pd
import polars as pl
import pyarrow as pa
import matplotlib.pyplot as plt
import matplotlib.animation as animation

import io
```

```python
import plotly.express as px
from IPython.display import clear_output # clear the previous plot

import plotly.io as pio
pio.renderers.default = "notebook_connected"

result = run_thread("q1",1, "../../resources/tpc-h/data/scale=1/partition=10/parquet")

frame = 0
df = pd.DataFrame()

for record in result:
    buf = io.StringIO(record)
    tmp = pl.read_json(buf)
    df_temp = tmp.to_pandas()
    df_temp['return flag, line status'] = "(" + df_temp['l_returnflag'] + ', ' + df_temp['l_linestatus'] + ")"
    df_temp['frame'] = frame
    df = pd.concat([df, df_temp])
    df['average price'] = df['avg_price']
    clear_output(wait=True)
    fig = px.bar(
        df, x='return flag, line status', y='average price', animation_frame='frame', animation_group='return flag, line status', 
        range_y=[df_temp['avg_price'].min() - 500, df_temp['avg_price'].max()+ 500], color="return flag, line status", 
        title="average price grouped by (return flag, line status) NetId: kf14, yz96, yl136, siruiw4"
        )
    fig.show()
    frame += 1
```
