```python
from tpch_polars import run_thread
import numpy as np
import pandas as pd
import polars as pl
import pyarrow as pa
import matplotlib.pyplot as plt
import matplotlib.animation as animation

import io
```

```python
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "notebook_connected"

result = run_thread("q1",1, "../../resources/tpc-h/data/scale=1/partition=10")

for record in result:
    buf = io.StringIO(record)
    tmp = pl.read_json(buf)
    df = tmp.to_pandas()
    df['myX'] = "(" + df['l_returnflag'] + ', ' + df['l_linestatus'] + ")"
    plt.cla()
    plt.yscale('log')
    plt.bar(x = df['myX'], height = df['avg_price'])
    plt.show()
```
